```python
import pandas as pd
```

```python
yc_data = pd.read_json("yc_data.json")
```

```python
yc_data
```

```python
yc_data["Funding"] = yc_data["Funding"].apply(lambda x: 0 if not x else x)
```

```python
yc_data["Funding"] = yc_data["Funding"].apply(lambda x: x.replace("m", "").replace("$", "").replace(",","") if type(x) == str else x)
```

```python
yc_data["Funding"] = yc_data["Funding"].apply(lambda x: float(x) if type(x) == str else x)
```

```python
yc_data["Funding"]
```

```python
yc_data.to_json("yc_data_cleaned.json", orient='records')
```
